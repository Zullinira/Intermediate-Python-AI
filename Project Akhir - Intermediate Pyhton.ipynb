### 5. Get rid of the missing value

#### 5.a) Drop columns which has missing value that cannot be filled with any logical value (like name, host_id, etc)


```python
dataairbnb.drop(['name','host_name'], axis=1, inplace=True)
```

```python
#kolom dan null baru
dataairbnb.isna().sum()
```

kolom name dan host_name menjadi hilang

#### 5.b) Fill the missing value in last_review with the most common date


```python
dataairbnb['last_review'].mode()
```

```python
dataairbnb['last_review']=dataairbnb['last_review'].fillna(dataairbnb['last_review'].mode()[0])
```

```python
display(dataairbnb)
```

```python
#cek nan pada last review
```

```python
dataairbnb.isna().sum()
```

berdasarkan data diatas bisa dilihat bahwa kolom last_review sudah tidak memiliki nan lagi

#### 5.c) Fill the missing value in reviews_per_month with the mean


```python
#nilai mean pada reviews_per_month
dataairbnb['reviews_per_month'].mean()
```

```python
#mengisi missing value dengan mean
dataairbnb['reviews_per_month']=dataairbnb['reviews_per_month'].fillna(dataairbnb['reviews_per_month'].mean())
```

```python
#cek data dan tabel
dataairbnb.isna().sum()
```

Nan pada 'reviews_per_month' telah menjadi 0

### 6. Make category based on the price: Cheap, Medium, and Expensive


```python
#cek tabel price
display(dataairbnb['price'])
```

```python
dataairbnb['price'].max()
```

```python
dataairbnb['price'].min()
```

```python
dataairbnb['price'].median()
```

```python
#price category
#kategori diambil dari data interquartile
#Expensive >175 --> Q3 (75%)
#Medium 69 - 175 --> Q1 - Q3
#Cheap <69 --> Q1 (25%)

dataairbnb['price category']=np.where(dataairbnb['price'] > 175,'Expensive',(np.where(dataairbnb['price']>69,'Medium','Cheap')))
```

### 8.  Visualize the price to compare the price of available rooms


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#menggunakan scatter plot 
plt.figure(figsize=(12,8))

plt.scatter(dataairbnb['price'],dataairbnb['availability_365'])

plt.title('Relationship Between Price and Room Availability', fontdict={'fontsize':20})
plt.xlabel('Price ($)',fontdict={'fontsize':15})
plt.ylabel('Room Availability',fontdict={'fontsize':15})

plt.xlim(-5,12000)
plt.ylim(-5,370)

plt.show()
```

```python
#price category
#kategori diambil dari data interquartile
#Expensive >175 --> Q3 (75%)
#Medium 69 - 175 --> Q1 - Q3
#Cheap <69 --> Q1 (25%)

dataairbnb['colors category']=np.where(dataairbnb['price'] > 175,3,(np.where(dataairbnb['price']>69,2,1)))
```

```python
dataairbnb
```

```python
#menggunakan scatter plot 
plt.figure(figsize=(12,8))

plt.scatter(dataairbnb['price'],dataairbnb['availability_365'],c=dataairbnb['colors category'])

plt.title('Relationship Between Price Category and Room Availability', fontdict={'fontsize':20})
plt.xlabel('Price ($)',fontdict={'fontsize':15})
plt.ylabel('Room Availability',fontdict={'fontsize':15})

plt.xlim(-5,12000)
plt.ylim(-5,370)


plt.show()
```
